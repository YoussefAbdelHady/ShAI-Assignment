## Task1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from colorama import Fore, Style

df = pd.read_csv('Salaries.csv')

# its a built in func to get length
no_rows = len(df)
no_columns = len(df.columns)
print('Number of rows = '+ str(no_rows))
print('Number of columns = '+str(no_columns)+'\n')

print('---------------------------------------------------------')


DTypes = df.dtypes
print(f'{Fore.RED}\nData Types Of Each Column :- {Style.RESET_ALL} '+'\n'+str(DTypes)+'\n')

print('---------------------------------------------------------')

# isna is a func its a shorcut for is Na--->null to check null values
# assumed that 0 is not a missed value & missed value is only empty particular column    
Missing_Values=df.isna().sum()
print(f'{Fore.RED}\nMissing values in each column:{Style.RESET_ALL}'+'\n'+str(Missing_Values))
```

## Task 2

```python
# Calculate basic statistics
# basic_statistics = df['TotalPay'].describe()
# print('Descriptive Statistics for Salary:'+'\n'+str(basic_statistics))

# Calculate mean
mean= df['TotalPay'].mean()
print('Mean = ', mean,'\n')

# Calculate median
median = df['TotalPay'].median()
print('Median = ',median,'\n')

# Calculate mode
mode = df['TotalPay'].mode().iloc[0]
print('Mode = ', mode, '\n')

# Calculate minimum
min= df['TotalPay'].min()
print('Min = ', min, '\n')

# Calculate minimum
max= df['TotalPay'].max()
print('Max = ', max, '\n')

# Calculate range
salary_range = df['TotalPay'].max() - df['TotalPay'].min()
print('Range of Salary = ', salary_range, '\n')

# Calculate standard deviation
std_deviation = df['TotalPay'].std()
print('Standard Deviation of Salary = ', std_deviation)
```

## Task3

```python
df = df.drop(columns= 'Status')
df = df.drop(columns= 'Notes')
df = df.drop(columns= 'Id')
df = df.drop(columns= 'Benefits')
df
```

```python
df = df.drop_duplicates()
df
```

```python
for x in df.index:
    if df.loc[x, 'JobTitle'] == 'Not provided':
        df.drop(x, inplace=True)
df
```

```python
def capitalize_name(name):
    return ' '.join(word.capitalize() for word in name.split())

# Find the index of the "employeeName" column
employee_name_index = df.columns.get_loc("EmployeeName")

# Capitalize the names in the "employeeName" column
df["EmployeeName"] = df["EmployeeName"].apply(capitalize_name)
df
```

```python
df["JobTitle"] = df["JobTitle"].apply(capitalize_name)
df
```

```python
for x in df.index:
    if df.loc[x, 'TotalPay'] == 0.0:
        df.drop(x, inplace=True)
plt.show()
df
```

```python
df.to_csv('modifiedd.csv', index=False)
```

```python
df["BasePay"] = df["BasePay"].replace('','0',regex=True)
df["OvertimePay"] = df["OvertimePay"].replace('','0',regex=True)
df["OtherPay"] = df["OtherPay"].replace('','0',regex=True)
df["TotalPay"] = df["TotalPay"].replace('','0',regex=True)
df["TotalPayBenefits"] = df["TotalPayBenefits"].replace('','0',regex=True)
df.to_csv('modifiedtask2.csv', index=False)
df
```

## Task 4

```python
salary_data = df['TotalPay']

# Create a histogram
plt.hist(salary_data, bins=15, color='blue', edgecolor='black')
# Add labels and title
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('Distribution of Salaries')
# Show the plot
plt.show()
```

```python
plt.figure(figsize=(8, 8))
df['JobTitle'].value_counts().nlargest(14).plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Employees in Different Departments')
plt.show()
```

```python
# Create groups of 50 rows each
group_size = 50
df['group'] = (df.index // group_size) + 1

# Calculate summary statistics for each group
grouped_stats = df.groupby('group')['TotalPay'].agg(['count', 'mean', 'median', 'std'])

# Print the summary statistics
print(grouped_stats)

# Compare average salaries across different groups
average_salaries = grouped_stats['mean']
print("Comparison of Average Salaries Across Groups:")
print(average_salaries)
```

```python
import seaborn as sns
TotalPayBenefits_column = 'TotalPayBenefits'  # Replace with the actual column name

# Check correlation between 'salary' and 'numerical_column'
correlation = df['TotalPay'].corr(df[TotalPayBenefits_column])
print(f"Correlation between 'TotalPay' and '{TotalPayBenefits_column}': {correlation}")

# Plot scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df[TotalPayBenefits_column], y=df['TotalPay'])
plt.title(f'Scatter Plot: Salary vs {TotalPayBenefits_column}')
plt.xlabel(TotalPayBenefits_column)
plt.ylabel('TotalPay')
plt.show()
```
